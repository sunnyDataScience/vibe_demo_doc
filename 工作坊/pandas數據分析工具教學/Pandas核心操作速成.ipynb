## <span style="color: #2E86AB; font-size: 1.4em;">步驟 1：讀取 (Read) - 載入資料</span>



**<span style="color: #06A77D; font-size: 1.1em;">口訣</span>**：<span style="color: #2E86AB; font-weight: bold;">萬物始於數據</span>





````python
# 導入必要的套件
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 設定中文字型（避免圖表中文亂碼）
plt.rcParams['font.family'] = 'Noto Sans CJK JP'
plt.rcParams['axes.unicode_minus'] = False  # 解決負號變方框問題

# 讀取 CSV 檔案
df = pd.read_csv('../../資料集/Taiwan_SuperMarket_Sales_2025.csv')

print("✅ 資料載入成功！")
print(f"資料形狀：{df.shape[0]} 筆記錄，{df.shape[1]} 個欄位")
````

## <span style="color: #F18F01; font-size: 1.4em;">步驟 2：查看 (View) - 了解資料結構</span>



**<span style="color: #06A77D; font-size: 1.1em;">口訣</span>**：<span style="color: #F18F01; font-weight: bold;">先看一眼，心中有數</span>





````python
# 查看前 5 筆資料
print("=== 前 5 筆資料 ===")
df.head()

# 查看資料基本資訊
print("\n=== 資料基本資訊 ===")
df.info()

# 查看數值欄位的統計摘要
print("\n=== 數值欄位統計 ===")
df.describe()
````

## <span style="color: #A23B72; font-size: 1.4em;">步驟 3：清洗 (Clean) - 處理缺失值</span>



**<span style="color: #E63946; font-size: 1.1em;">目的</span>**：<span style="color: #A23B72; font-weight: bold;">確保資料品質</span>





````python
# 檢查缺失值
print("=== 缺失值檢查 ===")
missing_count = df.isnull().sum()
print(missing_count[missing_count > 0])

# 如果有缺失值，可以用 dropna() 移除
# df = df.dropna()

# 轉換日期格式（如果需要）
df['Date'] = pd.to_datetime(df['Date'])

print("\n✅ 資料清洗完成！")
````

## <span style="color: #06A77D; font-size: 1.4em;">步驟 4：新增 (Add) - 建立衍生欄位</span>



**<span style="color: #E63946; font-size: 1.1em;">目的</span>**：<span style="color: #06A77D; font-weight: bold;">從現有資料創造新資訊</span>





````python
# 從日期中提取月份（如果還沒有）
if 'Month' not in df.columns:
    df['Month'] = df['Date'].dt.month

# 計算平均客單價（如果需要的話）
# df['Avg_Transaction'] = df['Sales'] / df['Quantity']

print("✅ 衍生欄位建立完成！")
print(f"\n新增欄位：Month")
df[['Date', 'Month', 'Sales']].head()
````

## <span style="color: #E63946; font-size: 1.4em;">步驟 5：分組 (Group) - 聚合分析</span>



**<span style="color: #06A77D; font-size: 1.1em;">口訣</span>**：<span style="color: #E63946; font-weight: bold;">分門別類，洞察規律</span>






**<span style="color: #2E86AB; font-size: 1.1em;">目標</span>**：



1. <span style="color: #2E86AB; font-weight: bold;">各商品線的總銷售額</span>



2. <span style="color: #2E86AB; font-weight: bold;">每月的總銷售額</span>



3. <span style="color: #2E86AB; font-weight: bold;">會員 vs 散客的總銷售額</span>





````python
# 1. 各商品線的總銷售額
product_sales = df.groupby('Product line')['Sales'].sum().sort_values(ascending=False)
print("=== 各商品線總銷售額 ===")
print(product_sales)

# 2. 每月的總銷售額
monthly_sales = df.groupby('Month')['Sales'].sum()
print("\n=== 每月總銷售額 ===")
print(monthly_sales)

# 3. 會員 vs 散客的總銷售額
customer_sales = df.groupby('Customer type')['Sales'].sum()
print("\n=== 會員 vs 散客總銷售額 ===")
print(customer_sales)
````

## <span style="color: #2E86AB; font-size: 1.4em;">步驟 6：排序 (Sort) - 找出關鍵資訊</span>



**<span style="color: #E63946; font-size: 1.1em;">目的</span>**：<span style="color: #2E86AB; font-weight: bold;">讓重要資訊一目了然</span>





````python
# 按銷售額排序（降序）
top_products = product_sales.sort_values(ascending=False)
print("=== 商品線銷售排名（前 5 名）===")
print(top_products.head())

# 也可以用 DataFrame 的方式排序
product_df = df.groupby('Product line').agg({
    'Sales': 'sum',
    'Invoice ID': 'count'  # 交易筆數
}).rename(columns={'Invoice ID': 'Transaction_Count'})
product_df = product_df.sort_values('Sales', ascending=False)

print("\n=== 商品線詳細統計 ===")
print(product_df.head())


````

## <span style="color: #A23B72; font-size: 1.4em;">步驟 7：視覺化 (Visualize) - 呈現洞察</span>



**<span style="color: #E63946; font-size: 1.1em;">目的</span>**：<span style="color: #A23B72; font-weight: bold;">用圖表說故事</span>





````python
# 設定圖表大小
plt.figure(figsize=(15, 5))

# 圖表 1：各商品線銷售額（橫向長條圖）
plt.subplot(1, 3, 1)
product_sales.plot(kind='barh', color='steelblue')
plt.title('Sales by Product Line', fontsize=14, fontweight='bold')
plt.xlabel('Sales (NT$)', fontsize=12)
plt.ylabel('Product Line', fontsize=12)
plt.tight_layout()

# 圖表 2：每月銷售趨勢（折線圖）
plt.subplot(1, 3, 2)
monthly_sales.plot(kind='line', marker='o', linewidth=2, markersize=8, color='coral')
plt.title('Sales Trend by Month', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales (NT$)', fontsize=12)
plt.xticks([1, 2, 3], ['Jan', 'Feb', 'Mar'])
plt.grid(True, alpha=0.3)

# 圖表 3：會員 vs 散客銷售占比（圓餅圖）
plt.subplot(1, 3, 3)
customer_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, 
                     colors=['steelblue', 'lightcoral'])
plt.title('Sales by Customer Type', fontsize=14, fontweight='bold')
plt.ylabel('')  # 移除 y 軸標籤

plt.tight_layout()
plt.show()

print("✅ 視覺化完成！")
````
